```python
import sys
import cv2
import matplotlib.pyplot as plt
from PIL import Image as im
import os
import numpy as np
from scipy.spatial import KDTree
from webcolors import hex_to_rgb
from webcolors import CSS3_HEX_TO_NAMES
import pandas as pd
```

```python
def convert_rgb_to_names(rgb_tuple):
    
    # a dictionary of all the hex and their respective names in css3
    css3_db = CSS3_HEX_TO_NAMES
    names = []
    rgb_values = []
    for color_hex, color_name in css3_db.items():
        names.append(color_name)
        rgb_values.append(hex_to_rgb(color_hex))
    
    kdt_db = KDTree(rgb_values)
    distance, index = kdt_db.query(rgb_tuple)
    return f'{names[index]}'
```

```python
color_list = []
file_name_list = []

path = '/content/feminine/'
file_list = os.listdir(path)

for file_name in file_list:
  img = cv2.imread('/content/feminine/'+file_name)
  print(file_name)
  
  cv2.waitKey()
  cv2.destroyAllWindows()
  # img = cv2.copyTo(src, mask)
  # plt.imshow(img)
  data = im.fromarray(img)
  data = np.array(data)
  x = data.shape[0]
  y = data.shape[1]

  data_list = []

  for i in range(x):
    for j in range(y):
      if data[i][j][0] != 0 or data[i][j][1] != 0 or data[i][j][2] != 0:
        data_list.append(data[i][j])
        # print(data[i][j])


  data_color = []

  for i in range(len(data_list)):
    r = data_list[i][0]
    g = data_list[i][1]
    b = data_list[i][2]
    color = convert_rgb_to_names((r,g,b))
    data_color.append(color)

  colordf = pd.DataFrame(data_color, columns=['COLOR'])
  color_info = colordf['COLOR'].value_counts()  
  color_info_df = pd.DataFrame(color_info)
  color_info_df = color_info_df.transpose()
  color_name = color_info_df.columns[0]

  color_list.append(color_name)
  file_name_list.append(file_name)

  print(color_name)

final = pd.DataFrame([file_name_list, color_list])
final = final.transpose()

final.to_csv('/content/drive/MyDrive/team/color.csv', index=None)
```
